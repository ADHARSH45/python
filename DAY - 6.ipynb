```python
import pandas as pd
ipl = pd.read_csv("IPL Matches 2008-2020.csv")
ipl.head(10)
```

```python
a = []
for _,row in ipl.iterrows():
    if row['winner'] == row['team1']:
        winner = row['team1']
        loser = row['team2']
    else:
        winner = row['team1']
        loser = row['team2']
    a.append((winner,loser))

df = pd.DataFrame(a,columns = ['winner','loser'])
df.head(20)
```

```python
team = df.groupby(['winner','loser']).size().reset_index(name = 'wins')
team.head(20)
```

 # Histograms

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.random.randn(10)
print(data)

plt.hist(data,color = 'g',edgecolor = 'y',bins = 5)

plt.xlabel("Values")
plt.ylabel("Frequency")
plt.title("Histogram example")

plt.show()
```

```python
x = [10,5,6,3,2]
y = [20,10,12,7,3]

plt.scatter(x,y,color = 'r')
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Scatter plot example")
plt.show()
```

```python
# pie plot example
categories = ["Housing","Transportation","Food","Entertainment","Utilities"]
expenses = [1200,300,400,200,150]

plt.figure(figsize = (5,16))

plt.pie(expenses,labels = categories,autopct = '%2.2f%%')
plt.legend()
plt.title("Monthly expenses")
plt.show()
```

```python
# box plot

data = [12,13,15,10,15,18,20,22,23,25,27,30,100,-34]
plt.boxplot(data)
plt.title("Boxplot exmaple with sample data")
plt.xlabel("data")
plt.ylabel("values")

plt.show()
```

```python
# stack plot
quarters = ['Q1','Q2','Q3','Q4']
electronics = [10000,12000,11000,10500]
clothing = [25000,6000,7500,9000]
home_applicances = [13000,7500,8200,9000]

plt.figure(figsize = (10,6))
plt.stackplot(quarters,electronics,clothing,home_applicances,labels= ['electronics','clothing','home applicances'])
plt.xlabel("Quarters")
plt.ylabel("Sales")
plt.title("Product category sales over quarters")
plt.legend()
plt.show()
```

```python
# 3d scatter plot
from mpl_toolkits.mplot3d import Axes3D

# generate data
x = np.linspace(-5,5,100)
y = np.linspace(-5,5,100)
x,y = np.meshgrid(x,y)
z = np.sin(np.sqrt(x**2 + y**2))

fig = plt.figure()
ax = fig.add_subplot(111,projection = '3d')
ax.scatter(x,y,z,c = z,cmap='viridis')
ax.set_xlabel('x axis')
ax.set_ylabel('y axis')
ax.set_zlabel('x axis')
plt.show()
```

```python
# subplot

x = np.linspace(0,10,100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.exp(-x)

# create a 2x2 plot
fig,axs = plt.subplots(2,2,figsize = (10,10))
axs[0,0].plot(x,y1,'r')
axs[0,0].set_title('sine waves')
axs[0,1].plot(x,y2,'g')
axs[0,1].set_title("cosine waves")
axs[1,0].plot(x,y3,'b')
axs[1,0].set_title("tan waves")
axs[1,1].plot(x,y4,'black')
axs[1,1].set_title('deccay exponential')

plt.tight_layout()
plt.show()
```
