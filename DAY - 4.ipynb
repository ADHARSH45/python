# iterator and generators
### iterator is an object thet allows you to traverse all the elements of a collecion once at a time
### it implements __iter__() and __next__() function
### use stopiteration exception to stop

```python
class Mynumber():
    def __iter__(self):
        self.num = 1
        return self

    def __next__(self):
        if self.num <=5:
            x = self.num
            self.num += 1
            return x
        else:
            raise StopIteration

mynumber = Mynumber()
myiter = iter(mynumber)
print(next(myiter))
for i in myiter:
    print(i)
```

```python
class Number:
    def __init__(self,stop,start = 0,step = 1):
        self.start = start
        self.stop = stop
        self.step = step
    def __iter__(self):
        self.num = self.start - 1
        return self
    def __next__(self):
        if self.num < self.stop:
            x = self.num
           
            self.num += self.step
            return self.num
            
        else:
            raise StopIteration




for i in Number(start = 10,stop = 20):
    print(i)
```

# Generators
### special type of iterator that defined using yeild keyword
### generate value one at a time
### more memory efficient 


```python
def my_generator():
    for i in range(1,6):
        yield i

gen = my_generator()
for value in gen:
    print(value)
```

# Iterator vs Generator
### iterator : require class with __iter__() and __next__()
### generator:  uses function with yield keyword
### iterator:  more code manual implementation
### generator : simple and concise implementation
### iterator : slightly more control
### Generator : more memory efficient

```python
ch  = "AABAABAAB"
res = []
i = 0
k = 3
while i < len(ch):
    res.append("".join(list(set(ch[i:i+k]))))
    i = i + k
print(res)
    
```
